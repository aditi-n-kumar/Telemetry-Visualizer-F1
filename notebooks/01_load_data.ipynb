```python
import fastf1
from fastf1 import plotting
import matplotlib.pyplot as plt

# Enable cache for faster repeated access
fastf1.Cache.enable_cache('../fastf1cache')

# Load Qualifying session of 2023 British GP
session = fastf1.get_session(2023, 'British Grand Prix', 'Q')

# Load session data
session.load()

# Pick driver (e.g., Lewis Hamilton)
lap = session.laps.pick_driver('HAM').pick_fastest()

# Get telemetry for that lap
tel = lap.get_telemetry()

# Plot speed over distance
plt.plot(tel['Distance'], tel['Speed'])
plt.xlabel('Distance (m)')
plt.ylabel('Speed (km/h)')
plt.title('Lewis Hamilton - 2023 British GP Q - Fastest Lap Speed')
plt.show()
```

```python
import fastf1
fastf1.Cache.enable_cache('../fastf1cache')

year = 2024
race_names = [
    'Bahrain Grand Prix', 'Saudi Arabian Grand Prix', 'Australian Grand Prix',
    'Azerbaijan Grand Prix', 'Miami Grand Prix', 'Monaco Grand Prix',
    'Spanish Grand Prix', 'Canadian Grand Prix', 'Austrian Grand Prix',
    'British Grand Prix', 'Hungarian Grand Prix', 'Belgian Grand Prix',
    'Dutch Grand Prix', 'Italian Grand Prix', 'Singapore Grand Prix',
    'Japanese Grand Prix', 'Qatar Grand Prix', 'US Grand Prix',
    'Mexico City Grand Prix', 'Las Vegas Grand Prix', 'Abu Dhabi Grand Prix'
]

sessions_to_download = ['Q', 'R']

for gp in race_names:
    for session_name in sessions_to_download:
        try:
            print(f"Loading {gp} {session_name} {year} ...")
            session = fastf1.get_session(year, gp, session_name)
            session.load()
            print(f"Loaded {gp} {session_name} successfully.")
        except Exception as e:
            print(f"Error loading {gp} {session_name}: {e}")
```

```python
import fastf1
from fastf1 import get_event_schedule
import pandas as pd

# Enable caching to store downloaded data
fastf1.Cache.enable_cache('../  fastf1cache')

# Get the 2024 event schedule
schedule = get_event_schedule(2024)

# Keep only Qualifying and Race sessions
saved_sessions = []

for _, row in schedule.iterrows():
    for sess in ['Q', 'R']:  # Qualifying and Race
        try:
            session = fastf1.get_session(2024, row['EventName'], sess)
            session.load(telemetry=False, weather=False)
            saved_sessions.append({
                'EventName': row['EventName'],
                'Session': sess,
                'RoundNumber': row['RoundNumber'],
                'EventDate': row['EventDate']
            })
            print(f"✔ Loaded: {row['EventName']} - {sess}")
        except Exception as e:
            print(f"✘ Skipped: {row['EventName']} - {sess} → {e}")

# Convert to DataFrame
df_sessions = pd.DataFrame(saved_sessions)

# Save to CSV
df_sessions.to_csv('2024_Qualifying_and_Race_Sessions.csv', index=False)
print("\n✅ Saved to 2024_Qualifying_and_Race_Sessions.csv")
```

```python
import fastf1
from fastf1 import plotting
import pandas as pd

# Enable caching
fastf1.Cache.enable_cache('../fastf1cache')

# Set year and sessions
year = 2024
sessions_to_download = ['Q', 'R']  # Qualifying and Race

# Master lists for storing all data
summary_data = []

for round_number in range(1, 25):  # Assume max 24 rounds
    for session_type in sessions_to_download:
        try:
            session = fastf1.get_session(year, round_number, session_type)
            session.load()
            
            print(f"Loaded: {session.event['EventName']} - {session_type}")
            
            # Basic metadata
            event_name = session.event['EventName']
            session_date = session.event['EventDate']
            session_name = session.name
            
            laps = session.laps

            for _, lap in laps.iterrows():
                summary_data.append({
                    'Event': event_name,
                    'Date': session_date,
                    'Round': round_number,
                    'Session': session_type,
                    'Driver': lap['Driver'],
                    'DriverNumber': lap['DriverNumber'],
                    'Team': lap['Team'],
                    'LapNumber': lap['LapNumber'],
                    'LapTime': lap['LapTime'],
                    'Compound': lap['Compound'],
                    'TyreLife': lap['TyreLife'],
                    'TrackStatus': lap['TrackStatus'],
                    'IsPersonalBest': lap['IsPersonalBest'],
                    'IsAccurate': lap['IsAccurate']
                })
        
        except Exception as e:
            print(f"Skipped Round {round_number} Session {session_type}: {e}")

# Convert to DataFrame
df = pd.DataFrame(summary_data)

# Save to Excel
output_file = "F1_2024_Data.xlsx"
df.to_excel(output_file, index=False)

print(f"\n✅ Saved all data to {output_file}")
```
